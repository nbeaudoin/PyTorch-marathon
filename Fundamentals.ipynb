# Matrix multiplication

Two methods in PyTorch
1. Element-wise multiplication
2. Matric multiplication (dot product)

Two rules that matrix multiplication need to follow:

1. Inner dimensions need to match
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work

2. The resulting matrix has the shape of the **outer dimensions**
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2)` @ `(3, 3)` -> `(3, 3)`

```python
# Element wise multiplication
tensor = torch.Tensor([1, 2, 3])
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

```python
### Common error in deep learning
tensor_A = torch.rand(3, 2)
tensor_B = torch.rand(3, 2)

torch.matmul(tensor_A, tensor_B)
```

```python
### Shape errors
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])

torch.mm(tensor_A, tensor_B)

```

We can fix this by using the transpose

```python
tensor_B.T, tensor_B.T.shape
```

```python
tensor_B, tensor_B.shape
```

```python
# The matrix multiplication operations works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <-- inner dimensions must math")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
torch.matmul(tensor_A, tensor_B.T)
```
